```python
import numpy as np

# n√∫mero de ret√¢ngulos
n = 100

# limites
a = 1
b = 2

# largura dos ret√¢ngulos
dx = (b - a) / n

# pontos de amostragem (Riemann √† direita)
x = a + np.arange(1, n + 1) * dx

# fun√ß√£o
f = x**3

# soma dos ret√¢ngulos
integral_num = np.sum(f * dx)

print("N√∫mero de ret√¢ngulos:", n)
print("Aproxima√ß√£o num√©rica da integral:", integral_num)
```

```python
import sympy as sp
sp.init_printing(use_latex=True)

# Defini√ß√£o da vari√°vel
x = sp.Symbol('x')

# Fun√ß√£o do exerc√≠cio
f = x**3

# Mostrando a f√≥rmula da integral
display(sp.Integral(f, (x, 1, 2)))

# Mostrando a antiderivada
F = sp.integrate(f, x)
display(sp.Eq(sp.Function('F')(x), F))

# Calculando a integral anal√≠tica
resultado = sp.integrate(f, (x, 1, 2))
display(sp.Eq(sp.Integral(f, (x, 1, 2)), resultado))
```

```python
import sympy as sp
sp.init_printing(use_latex=True)

# vari√°vel simb√≥lica
x = sp.Symbol('x')

# fun√ß√£o
f = x**3

print("A integral escrita assim:")
display(sp.Integral(f, (x, 1, 2)))

print("\n A antiderivada:")
F = sp.integrate(f, x)
display(sp.Eq(sp.Function('F')(x), F))

print("\n Aplica√ß√£o dos limites de 1 a 2:")
integral_aplicada = sp.Integral(f, (x, 1, 2)).doit()
display(
    sp.Eq(
        sp.Integral(f, (x, 1, 2)),
        sp.simplify(integral_aplicada)
    )
)

print("\n Resultado final:")
resultado_final = sp.Rational(15,4)
display(sp.Eq(resultado_final, float(resultado_final)))
```

**Gr√°fico com Ret√¢ngulos (Soma de Riemann)**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# N√∫mero de ret√¢ngulos (pode mudar para 20, 50, 100...)
n = 10

# Intervalo
a, b = 1, 2

# Largura dos ret√¢ngulos
dx = (b - a) / n

# Pontos usando Riemann √† esquerda
x = a + np.arange(n) * dx

# Fun√ß√£o
f = x**3

# Criar figura
fig, ax = plt.subplots(figsize=(8,5))

# Curva da fun√ß√£o x^3
xs = np.linspace(a, b, 500)
ax.plot(xs, xs**3, color='orange', linewidth=2)

# Desenhar ret√¢ngulos
for xi, fi in zip(x, f):
    rect = Rectangle((xi, 0), dx, fi, color='orange', alpha=0.6)
    ax.add_patch(rect)

# Ajustes do gr√°fico
ax.set_title("Aproxima√ß√£o Num√©rica da Integral de x¬≥ com Ret√¢ngulos")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.grid(alpha=0.3)
plt.show()
```
